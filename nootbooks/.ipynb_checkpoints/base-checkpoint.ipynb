## Show Image

```python
def show_img(disp_img):
    show_imgs([disp_img])

def show_imgs(disp_imgs, in_line = None, names = None):
    if not in_line:
        in_line = len(disp_imgs)
    
    while len(disp_imgs) % in_line != 0:
        disp_imgs += [None]
        
    nrows = int(len(disp_imgs) / in_line)
    
    fig = plt.figure(figsize=(6.4 * in_line, 4.8 * nrows))
    
    for i in range(len(disp_imgs)):
        ax = fig.add_subplot(nrows, in_line, i+1)
        ax.axis('off')
        if disp_imgs[i] is not None:
            if names:
                ax.set_title(names[i])
            ax.imshow(disp_imgs[i])
    plt.show()
```

```python
colord_img_origin = cv2.imread(get_dataset_path("image.jpg"), cv2.IMREAD_COLOR)
colord_img_origin = cv2.cvtColor(colord_img_origin, cv2.COLOR_BGR2RGB)
colord_img = colord_img_origin / 255
colord_img.shape
```

```python
gray_img_origin = cv2.imread(get_dataset_path("image.jpg"), cv2.IMREAD_GRAYSCALE)
gray_img_origin = cv2.cvtColor(gray_img_origin, cv2.COLOR_BGR2RGB)
gray_img = gray_img_origin / 255
gray_img.shape
```

```python
show_imgs([colord_img_origin, colord_img, gray_img_origin, gray_img], 2)
```

```python
imgs = []
for d in [0, 1, 2]:
    img = colord_img.copy()
    img[:,:,(d+1)%3] = 0
    img[:,:,(d+2)%3] = 0
    imgs += [img]

for d in [0, 1, 2]:
    img = colord_img.copy()
    img[:,:,d] = 0
    imgs += [img]
    
show_imgs(imgs, 3)
```

## Point-based operations

```python
# Image negative

img = gray_img.copy()

show_imgs([img, 1 - img])
```

```python
# Power Law Transform

gammas = np.arange(0.11, 1.1, 0.1).tolist() + np.arange(2, 11, 1).tolist()
gammas = np.around(gammas, decimals=1).tolist()

c = 1

imgs = []
for g in gammas:
    img = gray_img.copy()
    img =  c * img ** g
    imgs += [img]

show_imgs(imgs, 4, names=gammas)
```

```python
# Power Law Transform

cs = np.arange(0.11, 1.1, 0.1)
cs = np.around(cs, decimals=1).tolist()

imgs = []
for c in cs:
    img =  c * gray_img ** 1
    imgs += [img]

show_imgs(imgs, 4, names=cs)
```

```python
# Power Law Transform

gammas = np.arange(0.25, 1.1, 0.25).tolist()
cs = np.arange(0.25, 1.1, 0.25).tolist()

imgs = []
names = []
for c in cs:
    for g in gammas:
        imgs += [c * gray_img ** g]
        names += ["c={} g={}".format(c, g)]

show_imgs(imgs, 4, names=names)
```

## Try to fix Image

```python
img = gray_img.copy()

A, C, G = 0.1, 0.1, 1

fig = plt.figure(figsize=(6.4 * 2, 4.8 * 3))

ax = fig.add_subplot(3, 2, 1)
unique, counts = np.unique(img, return_counts=True)
ax.bar(unique, counts, width=10 / len(unique), align='edge')

ax = fig.add_subplot(3, 2, 2)
ax.axis('off')
ax.imshow(img)

img = A + C * img ** G

ax = fig.add_subplot(3, 2, 3)
unique, counts = np.unique(img, return_counts=True)
ax.bar(unique, counts, width=10 / len(unique), align='edge')

ax = fig.add_subplot(3, 2, 4)
ax.axis('off')
ax.imshow(img)

m1, m2 = img.min() , img.max()
img = ((img - m1) / (m2 - m1))

ax = fig.add_subplot(3, 2, 5)
unique, counts = np.unique(img, return_counts=True)
ax.bar(unique, counts, width=10 / len(unique), align='edge')

ax = fig.add_subplot(3, 2, 6)
ax.axis('off')
ax.imshow(img)
plt.show()
```

```python
img = colord_img.copy()

A, C, G = 0.2, 0.33, 1

fig = plt.figure(figsize=(6.4 * 2, 4.8 * 3))

ax = fig.add_subplot(3, 1, 1)
ax.axis('off')
ax.imshow(img)

for dim in range(3):
    img[:,:, dim] = A + C * img[:,:, dim] ** G

ax = fig.add_subplot(3, 1, 2)
ax.axis('off')
ax.imshow(img)

for dim in range(3):
    m1, m2 = img[:,:, dim].min() , img[:,:, dim].max()
    img[:,:, dim] = (img[:,:, dim] - m1) / (m2 - m1)

ax = fig.add_subplot(3, 1, 3)
ax.axis('off')
ax.imshow(img)
plt.show()
```

```python
As = [0.1, 0.2, 0.3]
Cs = [0.1, 0.2, 0.3]
Gs = [0.5, 1, 2]

imgs = []
names = []

for A in As:
    for C in Cs:
        for G in Gs:
            img = colord_img.copy()
            for dim in range(3):
                img[:,:, dim] = A + (C * (img[:,:, dim] ** G))
                
            imgs += [img.copy()]
            names += ["A {}, {}, {}".format(A, C, G)]
            
            for dim in range(3):
                m1, m2 = img[:,:, dim].min() , img[:,:, dim].max()
                img[:,:, dim] = (img[:,:, dim] - m1) / (m2 - m1)
            
            imgs += [img.copy()]
            names += ["B {}, {}, {}".format(A, C, G)]

show_img(colord_img)
show_imgs(imgs, 6, names=names)
```
