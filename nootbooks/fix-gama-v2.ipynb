```python
import math
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from src.load_images import set_database, get_dataset_path
from src.show_images import show_img, show_imgs
```

```python
set_database("custom")
```

```python
colored_img_origin = cv2.imread(get_dataset_path("image.jpg"), cv2.IMREAD_COLOR)
colored_img_origin = cv2.cvtColor(colored_img_origin, cv2.COLOR_BGR2RGB)
colored_img = colored_img_origin / 255

gray_img_origin = cv2.imread(get_dataset_path("image.jpg"), cv2.IMREAD_GRAYSCALE)
gray_img_origin = cv2.cvtColor(gray_img_origin, cv2.COLOR_BGR2RGB)
gray_img = gray_img_origin / 255
```

```python
show_imgs([colored_img_origin, colored_img, gray_img_origin, gray_img], 2)
```

```python
img = gray_img.copy()

A, C, G = 0.1, 0.1, 2.0

lMin, lMax, lLevel = 0.0, 1.0, np.median(img)

fig = plt.figure(figsize=(6.4 * 2, 4.8 * 3))

ax = fig.add_subplot(3, 2, 1)
unique, counts = np.unique(img, return_counts=True)
ax.plot(unique, counts)
plt.xlim([0, 1])

ax = fig.add_subplot(3, 2, 2)
ax.axis('off')
ax.imshow(img)

img = A + C * img ** G

ax = fig.add_subplot(3, 2, 3)
unique, counts = np.unique(img, return_counts=True)
ax.plot(unique, counts)
plt.xlim([0, 1])

ax = fig.add_subplot(3, 2, 4)
ax.axis('off')
ax.imshow(img)

m1, m2 = img.min() , img.max()
img = (img - m1) / (m2 - m1)
img = img ** math.log(lLevel, np.median(img))
img = img * (lMax - lMin) + lMin

ax = fig.add_subplot(3, 2, 5)
unique, counts = np.unique(img, return_counts=True)
ax.plot(unique, counts)
plt.xlim([0, 1])

ax = fig.add_subplot(3, 2, 6)
ax.axis('off')
ax.imshow(img)
plt.show()
```

```python
img = colored_img.copy()

A, C, G = 0.1, 0.1, 2.0

lMin, lMax = 0.0, 1.0
lLevel = [np.median(img[:,:,0]), np.median(img[:,:,1]), np.median(img[:,:,2])]

fig = plt.figure(figsize=(6.4 * 2, 4.8 * 3))

ax = fig.add_subplot(3, 2, 1)
for dim in range(3):
    unique, counts = np.unique(img[:,:,dim], return_counts=True)
    ax.plot(unique, counts)
plt.xlim([0, 1])

ax = fig.add_subplot(3, 2, 2)
ax.axis('off')
ax.imshow(img)

img = A + C * img ** G

ax = fig.add_subplot(3, 2, 3)
for dim in range(3):
    unique, counts = np.unique(img[:,:,dim], return_counts=True)
    ax.plot(unique, counts)
plt.xlim([0, 1])

ax = fig.add_subplot(3, 2, 4)
ax.axis('off')
ax.imshow(img)

for dim in range(3):
    m1, m2 = img[:,:,dim].min() , img[:,:,dim].max()
    img[:,:,dim] = (img[:,:,dim] - m1) / (m2 - m1)
    img[:,:,dim] = img[:,:,dim] ** math.log(lLevel[dim], np.median(img[:,:,dim]))
    img[:,:,dim] = img[:,:,dim] * (lMax - lMin) + lMin

ax = fig.add_subplot(3, 2, 5)
for dim in range(3):
    unique, counts = np.unique(img[:,:,dim], return_counts=True)
    ax.plot(unique, counts)
plt.xlim([0, 1])

ax = fig.add_subplot(3, 2, 6)
ax.axis('off')
ax.imshow(img)
plt.show()
```
